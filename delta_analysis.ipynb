# Delta for Barrier Call Options
## Delta behaviors
### For up and out calls: 
1. S0 far from the barrier: behaves similarly to a **vanilla call** (positive, increase with S0)
2. S0 apporaches the barrier: starts to **decrease and become even negative** near the barrier due to the increased likelihood of knock-out
3. Beyond the barrier: **0**, since the option value is 0
### For up and in calls: 
1. S0 far from the barrier: **Near 0 but non-zero** probability of crossing the barrier before maturity. 
2. S0 apporaches the barrier: delta to **increase rapidly**, since small changes in S0 significantly impact the probability of hitting the barrier.
3. Beyond the barrier: aligns with the delta of a vanilla call
   
## Reasonable negative deltas near the barrier
For up and out calls: as the initial price (S0) approaches the barrier from below, the probability of hitting the barrier increases. This reduces the option's value because the payoff becomes less likely to materialize. This effect can cause the estimated delta to be negativeeven for a callbecause you're comparing prices at S0+h and S0-h and finding that the "higher" price scenario leads to more frequent knock-outs.
